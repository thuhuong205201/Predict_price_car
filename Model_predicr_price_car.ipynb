```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split
```

```python
df_sample = pd.read_csv("/content/drive/MyDrive/datapreparing/Final Exam 19 12/dataset/sample.csv")
df_test = pd.read_csv("/content/drive/MyDrive/datapreparing/Final Exam 19 12/dataset/test.csv", sep=";")
df_train = pd.read_csv("/content/drive/MyDrive/datapreparing/Final Exam 19 12/dataset/train.csv", sep=";")
```

```python
df_train["engine"] = df_train["engine"].str.rstrip(" CC")
df_train["max_power"] = df_train["max_power"].str.rstrip(" bhp")
df_train["mileage"] = df_train["mileage"].str.rstrip(" kmpl")
df_train["mileage"] = df_train["mileage"].str.rstrip(" km/kg")
```

```python
df_train['engine'] = df_train['engine'].astype(float)
#df_train['max_power'] = df_train['max_power'].astype(object)
df_train['mileage'] = df_train['mileage'].astype(float)
```

```python
print("Shape of train data Before dropping any Row: ",df_train.shape)

df_train = df_train[df_train['mileage'].notna()]
print("Shape of train data After dropping Rows with NULL values in Mileage: ",df_train.shape)
df_train = df_train[df_train['engine'].notna()]
print("Shape of train data After dropping Rows with NULL values in Engine : ",df_train.shape)
df_train = df_train[df_train['max_power'].notna()]
print("Shape of train data After dropping Rows with NULL values in Max_Power  : ",df_train.shape)
df_train = df_train[df_train['torque'].notna()]
print("Shape of train data After dropping Rows with NULL values in Torque  : ",df_train.shape)
df_train = df_train[df_train['seats'].notna()]
print("Shape of train data After dropping Rows with NULL values in Seats  : ",df_train.shape)
```

```python
df_train.info()
```

```python
df_train['max_power'] = df_train['max_power'].astype(float)
```

```python
df_train.isnull().sum()
```

```python
var = 'fuel'
print(df_train[var].value_counts())

Fuel_t = df_train[[var]]
Fuel_t = pd.get_dummies(Fuel_t,drop_first=True)
print(Fuel_t.head())

var1 = 'transmission'
print(df_train[var1].value_counts())

transmission_t = df_train[[var1]]
transmission_t = pd.get_dummies(transmission_t,drop_first=True)
print(transmission_t.head())

var2 = 'owner'
print(df_train[var2].value_counts())
owner_t = df_train[[var2]]
owner_t = pd.get_dummies(owner_t,drop_first=True)
owner_t.head()

var3 = 'seller_type'
print(df_train[var3].value_counts())
seller_t = df_train[[var3]]
seller_t = pd.get_dummies(seller_t,drop_first=True)
seller_t.head()

```

```python
df_train= pd.concat([df_train,Fuel_t, seller_t,transmission_t,owner_t],axis=1)
df_train.head()
```

```python
df_train.drop(["fuel"],axis=1,inplace=True)
df_train.drop(["transmission"],axis=1,inplace=True)
df_train.drop(["owner"],axis=1,inplace=True)
df_train.drop(["seller_type"],axis=1,inplace=True)
```

```python
df_train['fuel_Diesel'] = df_train['fuel_Diesel'].astype(float)
df_train['fuel_LPG'] = df_train['fuel_LPG'].astype(float)
df_train['fuel_Petrol'] = df_train['fuel_Petrol'].astype(float)
df_train['seller_type_Individual'] = df_train['seller_type_Individual'].astype(float)
df_train['seller_type_Trustmark Dealer'] = df_train['seller_type_Trustmark Dealer'].astype(float)
df_train['transmission_Manual'] = df_train['transmission_Manual'].astype(float)
df_train['owner_Fourth & Above Owner'] = df_train['owner_Fourth & Above Owner'].astype(float)
df_train['owner_Second Owner'] = df_train['owner_Second Owner'].astype(float)
df_train['owner_Test Drive Car'] = df_train['owner_Test Drive Car'].astype(float)
df_train['owner_Third Owner'] = df_train['owner_Third Owner'].astype(float)
```

```python
df_train.drop(["torque"],axis=1,inplace=True)
```

```python
Y = df_train['selling_price']
X = df_train.drop(columns=['selling_price', 'name'])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=9)

print('X train shape: ', X_train.shape)
print('Y train shape: ', Y_train.shape)
print('X test shape: ', X_test.shape)
print('Y test shape: ', Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, Y_train)
y_pred= linear_reg.predict(X_test)
print("Accuracy on Traing set: ",linear_reg.score(X_train,Y_train))
print("Accuracy on Testing set: ",linear_reg.score(X_test,Y_test))
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error

print("\t\tError Table")
print('MAE      : ', metrics.mean_absolute_error(Y_test, y_pred))
print('MSE      : ', metrics.mean_squared_error(Y_test, y_pred))
print('Root MSE : ', np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
print('R-SE     : ', metrics.r2_score(Y_test, y_pred))
```

```python
compare = pd.DataFrame({'Prediction': y_pred, 'Test Data' : Y_test, 'Abs error': abs(Y_test - y_pred), 'AAD%': abs(Y_test - y_pred)/Y_test*100})
compare.head(10)
```
